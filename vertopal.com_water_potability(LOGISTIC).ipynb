<p>imputer = SimpleImputer(strategy="mean") df[df.columns] =
imputer.fit_transform(df)</p>
<p>X = df.drop("Potability", axis=1)<br />
y = df["Potability"]</p>
<p>X_train, X_test, y_train, y_test = train_test_split( X, y,

)</p>
<p>scaler = StandardScaler()<br />
X_train = scaler.fit_transform(X_train) X_test =
scaler.transform(X_test)</p>
<p>model = LogisticRegression()<br />
model.fit(X_train, y_train)</p>
<p>LogisticRegression()</p>
<p>y_pred = model.predict(X_test)</p>
<p>accuracy = accuracy_score(y_test, y_pred)<br />
print("Accuracy:", accuracy)<br />
