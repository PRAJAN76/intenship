```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
dia=pd.read_excel("/content/drive/MyDrive/diabetes (1).xlsx")
dia.head
```

```python
dia.isnull().sum()
```

```python
ind = dia[['age','mass','insu','plas']]
dep = dia['class']
```

```python
Logr = LogisticRegression()
Logr.fit(ind,dep)
```

```python
age = int(input("Enter the age:"))
mass = int(input("Enter the mass:"))
insulin = int(input("Enter the insulin level："))
plasma = int(input("Enter the plasma level:"))
pred = Logr.predict([[age, mass, insulin,plasma]])
print(pred)
```

```python
Logr.score(ind,dep)
```
